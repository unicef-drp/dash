```python
import pandas as pd
import json
```

```python
# Load the excel file again with the correct headers
df = pd.read_excel("master_list.xlsx", sheet_name="Master List", header=1)

# Filter rows where 'TM Dashboard' column is 'Yes'
filtered_df = df[df['TM Dashboard'] == 'Yes']

# Remove trailing spaces from the 'Subdomain' column in code_df
filtered_df['Subdomain'] = filtered_df['Subdomain'].str.strip()

# Display the first few rows of the 
filtered_df
```

```python
# Load the excel file again with the correct headers
code_df = pd.read_excel("master_list.xlsx", sheet_name="Subdomain Descriptions")

# Remove trailing spaces from the 'Subdomain' column in code_df
code_df['Subdomain'] = code_df['Subdomain'].str.strip()

code_df['Subdomain'] = code_df['Subdomain'].str.replace('^[0-9]*\.[0-9]*\.\s', '', regex=True)

code_df.head()
```

```python
code_df['Subdomain'].unique()
```

```python
# Now merge using the new column and Subdomain in code_df
merged_df = pd.merge(filtered_df, code_df, left_on='Subdomain', right_on='Subdomain', how='left')

# Rename the column from 'Dashboard_Code' to 'CODE'
merged_df.rename(columns={'Dashboard_Code': 'CODE'}, inplace=True)

merged_df[['Subdomain', 'CODE']].head()
```

```python
merged_df['Subdomain'].unique()
```

```python
# For each domain, list the associated subdomains
domain_dict = merged_df.groupby('Domain_x')['Subdomain'].unique().to_dict()

# For each subdomain, list the associated indicators and their codes
subdomain_dict = merged_df.groupby('Subdomain').apply(lambda x: {'name': x['Subdomain'].iloc[0], 'code': x['CODE'].iloc[0], 'indicators': x[['Indicator Name', 'Code']].to_dict(orient='records')}).to_dict()

# Combine the dictionaries
final_dict = {
    'domains': domain_dict,
    'subdomains': subdomain_dict
}

final_dict
```

```python
# Check which subdomains have 'nan' for the code
nan_subdomains = merged_df[merged_df['CODE'].isna()]['Subdomain'].unique()

# Display the subdomains with 'nan' codes
print("Subdomains with 'nan' codes:", nan_subdomains)
```

```python
# Convert any numpy arrays to lists
for domain, subdomains in final_dict['domains'].items():
    final_dict['domains'][domain] = list(subdomains)

# Save the final_dict as a JSON file
json_filename = "crm_framework_indicators.json"
with open(json_filename, "w") as outfile:
    json.dump(final_dict, outfile)

json_filename
```
